```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#sample Data
x = np.array([500,750,1000,1250,1500]).reshape(-1,1) #size of the house
y = np.array([150,200,250,300,350])   #price of the house
print(x)
print(y)
```

```python
#create and train model
model = LinearRegression()    # this is your algorithm
model.fit(x,y)   # learn the pattern hidden in x and y
```

```python
#make prediction
y_pred = model.predict(x)  #prediction is happening here
```

```python
# plot the data and the model
plt.scatter(x, y, color='blue', label='Actual')
plt.plot(x, y_pred, color = 'red', label= 'Predicted')
plt.xlabel('size of the house (sq ft)')
plt.ylabel('price of the house($)')
plt.title('linear regression model')
plt.legend()
plt.show()
```
